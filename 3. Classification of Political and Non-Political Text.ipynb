#### We have way too large a percentage of non-political titles.  Let's trim them down before we train.  

```python
polit_comments = comments[political].copy()
non_polit_comments = comments[-political][:1584].copy()

comments = pd.concat([polit_comments, non_polit_comments])
```

### Feature Prep for Classification

Let's first get our X (training/testing features) and Y (target labels) ready to go to begin creating and testing various text classification schemes for identifying whether or not a message written on Reddit is political in nature or not.  

```python
def encoder(string_of_text):
    return string_of_text.encode('utf-8').strip()

comments['encoded_comment'] = comments.comment.apply(encoder)

comments['is_political'] = (comments.category == 'POLITICAL').astype(int)

X = np.asarray(comments['encoded_comment'].copy(), dtype='|S6')
Y = np.asarray(comments.is_political.copy())
```

### Classification Using Multinomial Naive Bayes with Vectorized Text

```python
from sklearn.feature_extraction.text import CountVectorizer

import nltk.stem
english_stemmer = nltk.stem.SnowballStemmer('english')
class StemmedCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(StemmedCountVectorizer, self).build_analyzer()
        return lambda doc: (english_stemmer.stem(w) for w in analyzer(doc))
    
## Let' uncomment this later.  I want to see if we can get better results with probabilities


pipeline = Pipeline([
                     ('vectorizer', StemmedCountVectorizer(decode_error='replace')),
                     ('tfidf_transformer',  TfidfTransformer()),
                     ('classifier', MultinomialNB()) 
                     ])

pipeline.set_params(vectorizer__min_df=1, vectorizer__stop_words='english')


from sklearn.cross_validation import KFold


k_fold = KFold(n=len(X), n_folds=15, indices=False, shuffle=True)
scores = []


for train_indices, test_indices in k_fold:
    train_text = X[train_indices]
    train_y = Y[train_indices]

    test_text = X[test_indices]
    test_y = Y[test_indices]

    pipeline.fit(train_text, train_y)
    
    y_pred_proba = pipeline.predict_proba(test_text)
    # This is for predicting based on positive is_political
#     y_pred = (y_pred_proba[:,1] >= .675).astype(int)
    
    # This is for predicting on negative prob
    y_pred = (y_pred_proba[:,0] <= .48).astype(int)
    
    score = float(sum(y_pred == test_y))/len(test_y)
    
#     score = pipeline.score(test_text, test_y)
    scores.append(score)

score = sum(scores) / len(scores)
print(score)
```

```python
from scipy import interp


from sklearn.metrics import roc_curve, auc
from sklearn.cross_validation import StratifiedKFold


# Run classifier with cross-validation and plot ROC curves
cv = StratifiedKFold(Y, n_folds=6)
classifier = pipeline

mean_tpr = 0.0
mean_fpr = np.linspace(0, 1, 100)
all_tpr = []

for i, (train, test) in enumerate(cv):
    probas_ = classifier.fit(X[train], Y[train]).predict_proba(X[test])
    # Compute ROC curve and area the curve
    fpr, tpr, thresholds = roc_curve(Y[test], probas_[:, 1])
    mean_tpr += interp(mean_fpr, fpr, tpr)
    mean_tpr[0] = 0.0
    roc_auc = auc(fpr, tpr)
    pl.plot(fpr, tpr, lw=1, label='ROC fold %d (area = %0.2f)' % (i, roc_auc))

pl.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Luck')

mean_tpr /= len(cv)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
pl.plot(mean_fpr, mean_tpr, 'k--',
        label='Mean ROC (area = %0.2f)' % mean_auc, lw=2)

pl.xlim([-0.05, 1.05])
pl.ylim([-0.05, 1.05])
pl.xlabel('False Positive Rate')
pl.ylabel('True Positive Rate')
pl.title('Receiver operating characteristic example')
pl.legend(loc="lower right")
pl.show()
```

### Results of Political Classification

The attempt at simply using Naive Bayes coupled with Text Vectorization was ok.  We can do better.  

Similar to the approach we'll be using for detecting partisanship, we will use a Latent Dirichlet Allocation (LDA) method to create a topic model for each document.  Essentially, this acts as a method of dimensionality reduction for text.  

# LDA

### Step 1

Import all LDA utilities

```python
from declass.utils import gensim_helpers
from gensim import models, similarities, corpora
import gensim
```

### Step 2

Run the Reddit comments data through the LDA (gensim) algorithm.  Generate a dataframe of the respective topic scores for each comment.  

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
import nltk.stem
import string

from topic_modeling import df_tokenizer



stop = stopwords.words('english')
for i in list(string.punctuation):
    stop.append(i)
    
comments['tokens'] = comments.encoded_comment.apply(df_tokenizer)

# political = comments.category == 'POLITICAL'

dictionary = gensim.corpora.Dictionary(comments.tokens)
dictionary.filter_extremes()
dictionary.save('comment.dict')

corpus = [dictionary.doc2bow(text) for text in comments.tokens]
corpora.MmCorpus.serialize('data/comments_corpus.mm', corpus)

model = models.ldamodel.LdaModel(corpus, num_topics=500, id2word=dictionary)
"""
Random Forest Results

num_topics = 500, .48 threshold 
71.6% accuracy

num_topics = 100  .48 threshold
69.0% accuracy

num_topics = 200, .48 threshold
70.8% accuracy

Multinomial Naive Bayes Results

numtop = 500, .48 threshold
71% accuracy
            
"""

topics_df = gensim_helpers.get_topics_df(corpus, model)
```

### Step 3

Take the topic score attributes and use them as the features for ML Classification.  

```python
# topic_class = RandomForestClassifier()
topic_class = MultinomialNB()
topic_X = topics_df
```

```python
from sklearn.cross_validation import KFold


k_fold = KFold(n=len(topic_X), n_folds=15, indices=False, shuffle=True)
scores = []


for train_indices, test_indices in k_fold:
    train_text = topic_X[train_indices]
    train_y = Y[train_indices]

    test_text = topic_X[test_indices]
    test_y = Y[test_indices]
    
    topic_class.fit(train_text, train_y)
    
    y_pred_proba = topic_class.predict_proba(test_text)
    # This is for predicting based on positive is_political
#     y_pred = (y_pred_proba[:,1] >= .675).astype(int)
    
    # This is for predicting on negative prob
    y_pred = (y_pred_proba[:,0] <= .485).astype(int)
    
    score = float(sum(y_pred == test_y))/len(test_y)
    
#     score = pipeline.score(test_text, test_y)
    scores.append(score)

score = sum(scores) / len(scores)
print(score)
```
