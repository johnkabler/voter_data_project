```python
from declass.utils import gensim_helpers
from gensim import models, similarities, corpora
import pandas as pd
import numpy as np
import gensim

from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
import nltk.stem
import string

voters = pd.read_csv('data/makindo_data.csv')

#Comment the below line out when reading the cleaned data

voters = voters[['tweet', 'party_affiliation']]

# poli_tweets = pd.read_csv('data/navy/keyword-tweets.txt', sep='\t', header=None)
# poli_tweets.columns = ['political', 'tweet']
# political = poli_tweets.political == 'POLIT'
poli_comments = pd.read_pickle('data/reddit_poli_comments.pickle')


stop = stopwords.words('english')
for i in list(string.punctuation):
    stop.append(i)


def df_tokenizer(text): 
    token_list = []
    for sent in sent_tokenize(text):
        for word in word_tokenize(sent.lower()):
            if word not in stop:
                token_list.append(word)
    return token_list

poli_comments['tokens'] = poli_comments.comment.apply(df_tokenizer)
voters['tokens'] = voters.tweet.apply(df_tokenizer)

dictionary = gensim.corpora.Dictionary(poli_comments.tokens)
dictionary.filter_extremes()
dictionary.save('topic_dict.dict')

corpus = [dictionary.doc2bow(text) for text in voters.tokens]
corpora.MmCorpus.serialize('data/corpus.mm', corpus)

model = models.ldamodel.LdaModel(corpus, num_topics=100, id2word=dictionary)

topics_df = gensim_helpers.get_topics_df(corpus, model)
voters_topics = pd.concat([voters, topics_df], axis=1)

## This is a key output.  It is the list of voter tweets with topic scores as features.  

voters_topics.to_pickle('data/voters_topics_100.pickle')



#To aid in feature selection, let's provide a list of topics and their respective key words in pickle form
#for use in other tasks. 
topic_list = []
for t in xrange(model.num_topics):
    topic = {}
    topic['feature_name'] = 'topic_%s' % t
    topic['key_words'] = model.print_topic(t)
    topic_list.append(topic)
    
full_topics = pd.DataFrame(topic_list)

full_topics.to_pickle('data/full_topics.pickle')
```
